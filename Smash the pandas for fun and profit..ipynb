#### What is an heatmap?

In machine learning, a heatmap can be used to visualize the output of certain algorithms, particularly those involved in classification and clustering tasks. For instance:

- Confusion Matrix Heatmap: In classification tasks, a confusion matrix is used to evaluate the performance of a classifier. A heatmap representation of this matrix provides a visual indication of how well the classifier is performing by highlighting areas of correct and incorrect classifications.

- Feature Importance Heatmap: In predictive modeling, especially with techniques like decision trees or ensemble methods, feature importance measures are often computed to understand which features contribute the most to the model's predictions. A heatmap can be used to visualize these importance scores across different features.

- Correlation Heatmap: When working with datasets containing numerous features, it's essential to understand the relationships between these features. A correlation heatmap displays the pairwise correlations between features, making it easier to identify patterns or redundancies in the data.

- Activation Heatmap: In deep learning, particularly in convolutional neural networks (CNNs) used for image classification, activation heatmaps can be generated to visualize which parts of an input image are most important for the network's prediction. This can provide insights into how the model is making its decisions.

In all these cases, heatmaps serve as a visual aid to help understand the behavior of machine learning models and the underlying data patterns they capture.

This is a first solution a bit naive:
- read countries that belong to cio
- check the countries that won medals in 2008 in the index and make them unique
- foreach country that belong to cio if that is not in the 2008 index append to a list
- get the Name of no medal countries in loc.

```python
cio_countries = pd.read_csv('cio_countries.csv')
oo = pd.read_csv('olympics.csv')
edition_2008 = oo[(oo.Edition==2008)]
edition_2008.set_index('NOC', inplace=True)
edition_2008.sort_index()
unique_values = edition_2008.index.unique()
countries = len(unique_values)
loser_country = []
for country in cio_countries['CIOCode']:
    if country not in list(unique_values):
        loser_country.append(country)
        
indexed_data = cio_countries.set_index('CIOCode')
final = indexed_data.loc[loser_country].Country
```

This is much more better and concise solution:
   - read the two datasets
   - set an index on cio code in the cio country
   - merge the noc of medals with the first dataset
   - when it is not null you've a country who has not won anything in 2008.

```python
noc = pd.read_csv('cio_countries.csv')
oo = pd.read_csv('olympics.csv')
loc = oo[(oo.Edition==2008)]
noc.set_index('CIOCode', inplace=True)
noc.head()
medals_2008 = loc.NOC.value_counts()
noc['medals_2008'] = medals_2008
noc.head()
```

```python
loser_countries = noc[noc.medals_2008.isnull()]
loser_countries.head()
```

## GroupBy
GroupBy does three things:
- split a dataframe into groups based on some criteria.
- apply a function to each group indipendently.
- combine the result in dataframe.

Group by split a dataframe in groups, each group is a dataframe. It applies a function for each group and finally combines the results in a dataframe.

```python
oo = pd.read_csv('olympics.csv')
```

```python
edition_group = oo.groupby('Edition')
```

```python
type(edition_group)
```

```python
list(edition_group)
```

### Iteration throu the groups

```python
for key, value in oo.groupby('Edition'):
    print(key, value.City.iloc[0], type(value))
    
```

### On the Job
You might want .agg() aggregation function. Instruction are provided in the form of Python dictionary or a list. It allows to compute multiple statistics for each group in one shot.



```python
oo.groupby('Edition').size() # give the number of elements of the group
```

```python
oo = pd.read_csv('olympics.csv')
oo.groupby(['Edition','NOC','Medal']).agg(['min','max','count'])
```

```python
oo.groupby(['Edition','NOC','Medal']).agg(['count'])
```

```python
oo.groupby(['Edition','NOC','Medal']).agg({'Edition': ['min','max','count']})
```

```python
oo.loc[oo.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition': ['min','max', 'count']})
```

## Challange
- Using groupby, plot the total numbers of medals awarded at each of the games throught history
- create a list showing the total number of medals won for each country  over the history of the Olympics. For each country include the year of the first and the most recent Olympic medals wins.

```python
df = oo.groupby(['Edition']).agg({'Edition':['size']})
ax = df.plot(color='red', figsize=(10,3), kind='bar')
# Set labels
ax.set_xlabel("Edition")
ax.set_ylabel("Medals")
ax.legend(["Medals"])
```

```python
df = oo.groupby(['NOC']).agg({'NOC':['size'], 'Edition': ['min','max']})
df
```
